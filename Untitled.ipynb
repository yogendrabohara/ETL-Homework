```python
#import dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
```

```python
#store csv into dataframe
csv_file='./Resources/employee_reviews.csv'

employee_reviews=pd.read_csv(csv_file)
employee_reviews
```

```python
#dropping columns
employee_reviews.drop(columns=['location','dates','link','summary','helpful-count','pros','cons','advice-to-mgmt','job-title'], inplace=True)
```

```python
#Renaming columns
Employee_reviews=employee_reviews.rename(columns={'overall-ratings':'Overall',
                                'work-balance-stars':'WorkBalance',
                                'culture-values-stars':'CultureValues',
                                'carrer-opportunities-stars':'CareerOpportunities',
                                'comp-benefit-stars':'CompBenefit',
                                'senior-mangemnet-stars':'SeniorManagement'})
Employee_reviews.head()
```

```python
#connect to local database
rds_connection_string="postgres:postgres@127.0.0.1/Employee_reviews_db"
engine= create_engine(f'postgres://{rds_connection_string}')
```

```python
#Check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted Dataframe into database
Employee_reviews.to_sql(name='employee_reviews' ,con=engine , if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from Employee_reviews', con=engine).head()
```

```python
#Total Reviews for each corporation
total_reviews=Employee_reviews.groupby(["company"])["Overall"].count()
total_reviews


#Employee_reviews1 = Employee_reviews[Employee_reviews['CultureValues'] != 'none']
```

```python
#configure plot and ticks
total_reviews.plot(kind='bar', facecolor='blue', figsize=(10,5))
plt.title("Company Vs Reviews Count")
plt.ylabel("Total Reviews Count")
plt.xlabel("Company")
plt.show()
plt.savefig("./companyreviews.png")
```

```python
#company and overall rating 
com_Overall=Employee_reviews.groupby(['company'])['Overall'].mean()
com_Overall
```

```python
#configure plot and ticks
com_Overall.plot(kind='bar', facecolor='blue', figsize=(10,5))
plt.title("Company Vs Overall Rating")
plt.ylabel("Overall Rating")
plt.xlabel("Company")
plt.show()
plt.savefig("./Overallrating.png")
```

```python
#Creating Word Clouds on pros working with corporation

text = employee_reviews.pros.values
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'white',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (20, 10),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
plt.savefig("./pros.png")
```
